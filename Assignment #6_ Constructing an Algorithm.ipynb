## Problem #1: Strategy for Uncorrupted Robots
As an uncorrupted robot, the goal is to identify all corrupted robots with minimum use of the test bed. I propose these strategies to maximize exposure of corrupted robots:

1. **Pairwise Testing:** Initially, test robots in pairs. Since uncorrupted robots always tell the truth, any pair where both robots declare the other as uncorrupted is a pair of uncorrupted robots. We can safely exclude them from further tests.

2. **Handling Disagreements:** If a pair disagrees (one says the other is corrupted and vice versa), at least one of them must be corrupted. Keep these robots for further investigation.

3. **Cross Verification:** After initial pairing, take robots from the disagreement pairs and test them with robots confirmed as uncorrupted. This will help isolate who is corrupted among the disagreeing pairs.

4. **Iterative Process:** Repeat the cross-verification process, reducing the pool of potential corrupted robots each time.

5. **Optimization:** To minimize test bed usage, use combinatorial grouping strategies where possible, and avoid retesting pairs already confirmed as uncorrupted.

## Problem #2: Strategy for Corrupted Robots

As a corrupted robot, the objective is to maximize test bed usage and avoid detection. I propose the following strategies:

1. **Consistent Lying:** Always falsely accuse uncorrupted robots of being corrupted. This creates confusion and necessitates more tests. Since the corrupted robots can communicate with eachother this should be easy to coordinate.

2. **Collaboration:** Corrupted robots should agree on a common set of lies to maintain consistency in their accusations.

3. **Target Isolation:** Focus on accusing a specific subset of uncorrupted robots to create a scenario where it's hard to determine who is telling the truth without extensive cross-testing.

4. **Adaptation:** If the uncorrupted robots employ a systematic approach to testing, corrupted robots might need to adapt their strategy, possibly changing who they accuse over time to create further confusion.

## Problem #3: Minimum Number of Corrupted Robots for Indeterminate Status

The minimum number *K* of corrupted robots required to make it impossible to definitively prove which robot is corrupted depends on the total number of robots *N*. The critical point is when the number of corrupted robots is sufficient to create a scenario where their testimony can always counter the testimony of the uncorrupted robots, creating perpetual ambiguity.

1. If *K* is less than or equal to *N/2*, there's always a majority of uncorrupted robots, and their consistent truth-telling will eventually expose all corrupted robots.

2. If *K* is more than *N/2*, corrupted robots can always create a situation where it's impossible to distinguish between truthful and false testimonies without additional information.

Thus, the minimum number *K* such that it is impossible for any algorithm to definitively prove which robot is corrupted and which is not, is when *K>N/2*.
