###Traveling Salesman Problem (TSP) Solutions:

```python
# 5-Node matrix
from itertools import permutations

dist_matrix = [
    [0, 10, 15, 20, 25],
    [10, 0, 35, 25, 30],
    [15, 35, 0, 30, 20],
    [20, 25, 30, 0, 15],
    [25, 30, 20, 15, 0]
]
```

### Brute force approach:

---





```python
def calculate_total_distance(path):
    """Calculate the total distance of a given path"""
    total_dist = 0
    for i in range(len(path)):
        total_dist += dist_matrix[path[i-1]][path[i]]
    return total_dist

def brute_force_tsp(dist_matrix):
    """Solve TSP using brute force approach"""
    nodes = list(range(len(dist_matrix)))
    min_path = None
    min_dist = float('inf')

    for path in permutations(nodes):
        current_dist = calculate_total_distance(path)
        if current_dist < min_dist:
            min_dist = current_dist
            min_path = path

    return min_path, min_dist

# Solving TSP using brute force
brute_force_solution = brute_force_tsp(dist_matrix)
brute_force_solution
```

The brute force approach involves generating all possible permutations of the nodes and calculating
the total distance for each permutation. The shortest path is considered the solution which is the following:

> Shortest Path: (0, 1, 3, 4, 2) <br>
  Total Distance: 85







### Randomization algorithm:

---

```python
import random

def random_path(nodes):
    """Generate a random path"""
    path = nodes[:]
    random.shuffle(path)
    return path

def randomized_tsp(dist_matrix, iterations=10000):
    """Solve TSP using a randomization algorithm"""
    nodes = list(range(len(dist_matrix)))
    min_path = None
    min_dist = float('inf')

    for _ in range(iterations):
        path = random_path(nodes)
        current_dist = calculate_total_distance(path)
        if current_dist < min_dist:
            min_dist = current_dist
            min_path = path

    return min_path, min_dist

# Solving TSP using randomization
randomized_solution = randomized_tsp(dist_matrix)
randomized_solution
```

The randomization algorithm repeatedly generates random paths and keeps track of the shortest path found over a set number of iterations. This approach does not guarantee the optimal solution but can often find a good approximation in less time. When I ran the code multiple times I usually ended up with different solutions.
